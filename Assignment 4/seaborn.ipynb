```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#distplot

sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
#kind=scatter,reg,resid,kde,hex
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#pairplot
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
tips.describe
```

```python
#categorical plots

#barplot and count plot

sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
#boxplots and violinplot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
#stripplot and swarmplot
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
#factoplot

sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```

```python
#Matrix Plots
flights = sns.load_dataset('flights')
flights.head()
```

```python
#heatmap
sns.heatmap(tips.corr())
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
flightes_pvt = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(flightes_pvt)
```

```python
sns.heatmap(flightes_pvt,cmap='magma')
```

```python
sns.clustermap(flightes_pvt)
```
