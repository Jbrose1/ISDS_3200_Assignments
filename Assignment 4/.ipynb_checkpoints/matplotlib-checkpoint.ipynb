```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,color='purple')
```

```python
plt.plot(x,y,color='purple')
plt.xlabel('X label goes here')
plt.ylabel('Y label goes here')
plt.title('Title goes here')
plt.show()
```

```python
#nrows, ncolumns, plot number
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
#create plots using object oriented method
fig = plt.figure()

#add set of axes to the figure
axes = fig.add_axes([0.2,1,0.8,0.8])  #left,bottom,width, height (between 0 and 1)

#plot on the axes
axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y Label')
axes.set_title('Title goes here')

```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.2])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'r')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=3)
axes[0,0].plot(x,y)
axes[0,1].plot(x,y,'r')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
for ax in axes:
    ax.plot(x,y)
plt.tight_layout()
```

```python
#figure size and DPI

fig = plt.figure(figsize=(8,4),dpi=100)
```

```python
fig,axes  = plt.subplots(figsize=(12,3))
axes.plot(x,y)
```

```python
#saving plots to image file
fig.savefig("isds3200_2.png",dpi=400)
```

```python
#legends

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend()
```

```python
#location of legend

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3,label="x**3")
ax.legend(loc=1)
#1 upper right corner
#2 upper left corner
#3 lower left corner
#4 lower right corner
```

```python
#color
fig,ax = plt.subplots()
ax.plot(x,x**2,'g--')
ax.plot(x,x**3,'r.-')
ax.plot(x,y*2,color="#461d7c")
```

```python
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(x,x+1,color="#461d7c",linewidth=20.00)
ax.plot(x,x+2,color="#461d5c",linewidth=.25)
ax.plot(x,x+3,color="green",linestyle='--',marker='+',markersize=18)
```

```python

fig,ax = plt.subplots(1,3,figsize=(12,6))

ax[0].plot(x,x**2,x,x**3)

ax[1].plot(x,x**2,x,x**3)
ax[1].set_ylim([0,60])
ax[1].set_xlim([2,5])


ax[2].plot(x,x**2,x,x**3)
ax[2].axis('tight')
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,3))

axes[0].plot(x,x**2,x,x**3)
axes[0].grid(True)

axes[1].plot(x,x**2,x,x**3)
axes[1].grid(color='b',alpha=0.5,linestyle='dashed',linewidth=0.5)
```
