```python
#Question 9: Does the Breach type and location correlate to the number of individuals affected?
piv = df.pivot_table(values='Individuals_Affected',index='Type_of_Breach',columns='Location_of_Breached_Information')
piv.head(10)
```

```python
#Conclusion: There is no substantial correlation.
plt.figure(figsize = (10,5))
sns.heatmap(piv)
```
